```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv("iris.csv")
iris.plot(kind="scatter",x="SepalLengthCm",y="PetalLengthCm")
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv("iris.csv")
plt.figure(figsize=(10,3))
x=iris["SepalLengthCm"]
plt.hist(x,bins=30,color="green")
plt.xlabel("SepalLengthCm in cm")
plt.ylabel("count")
plt.show()
```
