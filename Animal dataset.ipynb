```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("class.csv")
plt.figure(figsize=(10,3))
x=data["Number_Of_Animal_Species_In_Class"]
plt.hist(x,bins=30,color="green")
plt.xlabel("Class_Number in cm")
plt.ylabel("count")
plt.show()
```
